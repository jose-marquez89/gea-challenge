## Presentation Graphic
A graphic representing the increase in accuracy across different approaches

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# arrange data
scores = [0.7413, 0.7788, 0.8079, 0.7407, 0.8230]
approaches = ["Baseline NNC", "10 Neighbors", 
              "30 Neighbors", "SVM", "RFC + NNC"]
```

```python
plt.style.use("seaborn")

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(approaches, scores, marker='o', color="green")
ax.grid(True)

plt.title("Accuracy Across 5 Approaches")
plt.ylim(.73, .83)
plt.savefig("accuracy.png")
```
